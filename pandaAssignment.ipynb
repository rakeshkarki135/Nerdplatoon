```python
import pandas as pd
```

```python
df = pd.read_csv("D:/project files/Nerdplatoon/pandas/file.csv")

df
```

```python
# pd.set_option('display.max_columns',19)
# pd.set_option('display.max_rows',200)

# remove the row with null columns on specified columns
df = df.dropna(subset = ['title','img_src'])

df.head(20)


```

```python
# to check the types of column values
# df['price'].apply(type)

df.loc[:,['price']] = df['price'].str.replace('$',' ', regex=False)

df
```

```python
df.loc[:,['price']] = df['price'].str.replace(',', '')
```

```python
df
```

```python
df.loc[:,['price_unit']] = df['price'].str.extract(r'(Monthly|Yearly)', expand=False)
```

```python
# change the type of data 
df.loc[:,['price']] = df['price'].str.replace(r'[^\d]','', regex=True).astype(int)
```

```python
df.loc[:,['price','price_unit']]
```

```python


df.rename(columns = {'interior':'interior_sq_ft'}, inplace=True)

df.columns

```

```python
print(df.loc[:,'interior_sq_ft'])

df.loc[:,'interior_sq_ft'] = df['interior_sq_ft'].str.replace(',','')

df.loc[:,'interior_sq_ft'] = df['interior_sq_ft'].str.replace(r'[a-zA-z\s\.]','',regex=True)
```

```python
# replacing nan value with 0
df.loc[:,'interior_sq_ft'] = df['interior_sq_ft'].fillna('0')
```

```python
df['interior_sq_ft'] = df['interior_sq_ft'].astype(int)
```
